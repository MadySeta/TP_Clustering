# 2  Clustering k-Means

## 2.1 Pour démarrer

```python
def k_means(path,name,k):
    datanp = get_data(path,name)

    # Run clustering method for a given number of clusters
    tps1 = time.time()
    model = cluster.KMeans(n_clusters=k, init='k-means++', n_init=1)
    model.fit(datanp)
    tps2 = time.time()
    
    return model, (round((tps2 - tps1)*1000,2))
```

```python
def print_Kmean(model,datanp,runtime,name):
    print("---------------------------------------")
    print("Affichage données initiales            "+ str(name))
    f0 = datanp[:,0] # tous les élements de la première colonne
    f1 = datanp[:,1] # tous les éléments de la deuxième colonne
    
    #plt.figure(figsize=(6, 6))
    plt.scatter(f0, f1, s=8)
    plt.title("Donnees initiales : "+ str(name))
    #plt.savefig(path_out+"Plot-kmeans-code1-"+str(name)+"-init.jpg",bbox_inches='tight', pad_inches=0.1)
    plt.show()
    
    print("------------------------------------------------------")
    print("Affichage après appel KMeans pour une valeur de k fixée")
    labels = model.labels_
    # informations sur le clustering obtenu
    iteration = model.n_iter_
    inertie = model.inertia_
    centroids = model.cluster_centers_
    
    #plt.figure(figsize=(6, 6))
    plt.scatter(f0, f1, c=labels, s=8)
    plt.scatter(centroids[:, 0],centroids[:, 1], marker="x", s=50, linewidths=3, color="red")
    plt.title("Données après clustering : "+ str(name) + " - Nb clusters ="+ str(model.n_clusters))
    #plt.savefig(path_out+"Plot-kmeans-code1-"+str(name)+"-cluster.jpg",bbox_inches='tight', pad_inches=0.1)
    plt.show()

    print("Résultats")
    print(" nb clusters =",model.n_clusters,"\n nb iter =",iteration, "\n inertie = ",inertie, "\n runtime = ", runtime,"ms")
```

## Xclara.arff

```python
xclara_name = 'xclara.arff'
model, runtime = k_means(path,xclara_name,3)
print_Kmean(model,get_data(path,xclara_name),runtime, xclara_name)
```

## 2.2 Application itérative de la méthode k-Means basée sur l’inertie

```python
name="xclara.arff"
inertia_score = []
for i in range(1,10):
    model, _ = k_means(path,name,i)
    inertia_score.append(model.inertia_)
inertia_score = [None] + inertia_score
plt.plot(inertia_score, label = "Inertie")
plt.title("Evolution de l'inertie en fonction du nombre de cluster k")
plt.xlabel('Nombre de cluster k')
plt.ylabel('Inertie')
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)
plt.xticks(range(len(inertia_score)))
plt.show()
```

Pour déterminer une bonne solution de clustering, il faut considérer le point d'inflexion. 
Pour cela, on peut utiliser la __méthode du coude__.

Ici, le point d'inflexion est lorsqu'on a un k=3.

## 2.3 Application itérative de la méthode k-Means basée sur des métriques d'évaluation externes

### Coefficient de silhouette

```python
square1_name = 'square1.arff'
silhouette_score_list = []

for i in range(2,20):
    model, t = k_means(path,square1_name,i)
    score = metrics.silhouette_score(get_data(path, square1_name), model.labels_)
    silhouette_score_list.append(score)

silhouette_score_list = [None, None] + silhouette_score_list
plt.plot(silhouette_score_list, label = "Inertie")
plt.title("Evolution de l'inertie en fonction  du nombre de cluster k")
plt.xlabel('Nombre de cluster k')
plt.ylabel('Coefficient de Silhouette')
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)
plt.xticks(range(len(silhouette_score_list)))
plt.show()
```

## 2.5 Version mini-batch de la méthode k-Means

```python
def mini_batch_k_means(path,name,k,bs):
    datanp = get_data(path,name)

    # Run clustering method for a given number of clusters
    tps1 = time.time()
    model = cluster.MiniBatchKMeans(n_clusters=k, batch_size=bs, random_state=42, n_init=1)
    model.fit(datanp)
    tps2 = time.time()
    
    return model, (round((tps2 - tps1)*1000,2))
```

### Diamond9.arff

```python
name = "diamond9.arff"
k = 9
model, runtime = mini_batch_k_means(path,name,k,2000)
print_Kmean(model,get_data(path,name),runtime, name)
```

```python
name = "diamond9.arff"
silhouette_score_list = []
bs = np.arange(1,2002,100)
for i in bs:
    model, t = mini_batch_k_means(path,name,9,i)
    score = metrics.silhouette_score(get_data(path, name), model.labels_)
    silhouette_score_list.append(score)

plt.plot(bs, silhouette_score_list)
plt.title("Evolution du coefficient de silhouette en fonction  du batchsize")
plt.xlabel('Batchsize')
plt.ylabel('Coefficient de Silhouette')
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)
plt.show()
```
